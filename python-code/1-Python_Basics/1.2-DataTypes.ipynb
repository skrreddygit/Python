### DataTypes

Data types are a classification of data which tel the compiler or interpreter how the programmer intend to use the data.
* They determine the type of operations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data.

```python
## Boolean datatype

print(type(True))
print(type(False))
print("=================")
is_true=True
is_false=False
print(type(is_true))
print(type(is_false))
```

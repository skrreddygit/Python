## logical Operators

```python
## AND,NOT,OR

x=True
y=True

result = x and y

print(result)
```

```python
x=True
y=False

result = x and y

print(result)
```

```python
## OR only if x and y are false then only it will give false remaining all the time we will get True

x=True
y=False

result = x or y

print(result)
```

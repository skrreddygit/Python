```python
## if statement

age = 18

if age>=18:
    print("You are elgible to vote")
```

```python
## else statement - this block will run if the if statement is False

age = 16

if age>=18:
    print("You are elgible to vote")
else:
    print("You are Not elgible to vote.since you are minor")
```

```python
## elif statement - this block allows us to check multiple conditions.

age = 17

if age<=13:
    print("You are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are elgible to vote")
```

```python
### nested Conditional Statements

# You can place one or more if,elif, or else statement inside another if,elif or else statement to create nested conditional statements

# Q) Number is even or odd or negative


num = int(input("Enter the number"))
print(f"Given Number is {num}")
if num>=0:
    print("Number is positive")
    if num%2 == 0:
        print("Given number is even number")
    else:
        print("Given number is odd number")
else:
    print("Given number is zero or negative")
```

```python
# Determine if a year is a leap year using nested condition statement

year = int(input("Enter the number"))

if year%4==0:
    if year%100 == 0:
        if year%400==0:
            print(f"{year} is leap year")
        else:
            print(f"{year} is not a leap year")
    else:
        print(f"{year} is a leap year")
else:
   print(f"{year} not a leap year")
```

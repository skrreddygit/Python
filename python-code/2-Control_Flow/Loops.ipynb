```python
## While loop -  it continues to execute as long as the condition is True

count = 0

while count<5:
    print(count)
    count=count+1

```

```python
count = 0
while count%2==0:
    print(count)
    count=count+1
```

```python
### loop Control Statements

## break - statement exits the loop prematurely

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## continue - statement skips the current iteration and continues with the next

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass = is a null operation; it does nothing

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested Loops - loop inside loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

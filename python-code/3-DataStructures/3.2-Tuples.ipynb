```python
## Immutable Nature of Tuples
## tuples are immutable, meaning their elements cannot be changed once assigned

lst = [1,2,3,4,5]
print(lst)

lst[1]="alien"
print(lst)
```

```python
num_tpl[1]="alien"
print(num_tpl)

## We cannot directly add the values in tuple. First change that to List then add and again change that to tuple
```

```python
## Tuple Methods
print(num_tpl)
print(num_tpl.count(3))
print(num_tpl.index(5)) # return the index of this value
```

```python
## Packing and Unpacking tuple

packed_tuple = 1,"Hello",3.14,"alien"
print(packed_tuple)

## Unpacking the tuple

a,b,c,d = 1, 'Hello', 3.14, 'alien'
print(a)
print(b)
print(c)
print(d)
```

```python
## unpacking using *

numbers = (1,2,3,4,5,6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
### Nested List
lst = [[1,2,3,4],[6,7,8,9],[1,"hello",3.14]]
print(lst[0])
print(lst[0][1])
print(lst[0][0:3])
```

```python
### Nested List
nested_tpl = ((1,2,3),(6,7,8),(1,"hello",3.14))
print(nested_tpl[0])
print(nested_tpl[0][1])
print(nested_tpl[1][2])
```

```python
# iterating over nested tuple

for sub_tpl in nested_tpl:
    for item in sub_tpl:
        print(item,end=" ")
    print()
```

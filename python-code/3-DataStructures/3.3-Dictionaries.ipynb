## Introduction to Dictionaries

* Dictionaries are unordered collections of items. 
* They store data in key-value pairs. 
* Keys must be unique and immutable (e.g., strings, numbers, or tuples),
* while values can be of any type.

```python
## Create a Dictionary

empty_dict = {}
print(empty_dict)
print(type(empty_dict))

empty_dict1 = dict()
print(empty_dict1)
```

```python
student = {"name":"alien","age":32,"grade": 4.25}
student
```

```python
## If the keys are same. the key will be replaced with the latest value
student = {"name":"alien","age":32,"name": 4.25}
student
```

```python
## Accessing the Dictionary Values
student = {"name":"alien","age":32,"grade": 4.25}
print(student['grade'])
print(student['age'])

## Accessing usinf get() method

print(student.get('grade'))
print(student.get("age"))
print(student.get("last_name","Not available"))
```

```python
## Modifying Dictionary Elements
## Dictionary are mutable,so you can add, update or delete

print(student)
```

```python
student["age"]=33 ## Updated the value for the key 'age'
print(student)
student['address'] = 'india' ## Added new key and value as india
print(student)
```

```python
student = {'name': 'alien', 'age': 33, 'grade': 4.25, 'address': 'india'}
del student['grade']  ## Deletes the key 'grade
```

```python
print(student)
```

```python
## Dictionary Methods to fecth the keys and Values

keys=student.keys()
values=student.values()

print(keys)
print(values)


items = student.items() ## to get all key value pairs
print(items)
print(type(items))
```

```python
## shallow copy 
student_copy = student
print(student_copy)
print(student)
```

```python
student["name"]="alien2"
print(student_copy)
print(student)
```

```python
# In shallow copy student_copy is also getting the same value as student once the student is updted. this shouldnt happen so we use

student_copy1=student.copy()
print(student_copy1)
```

```python
student["name"]="alien"
print(student_copy1)
print(student)
```

```python
## iterating Over Dictionaries
## We can use Loops to iterate over dictionaris,keys,values or items

## Iterating Over keys

for key in student.keys():
    print(key)

## Iterating Over Values

for value in student.values():
    print(value)

## Iterating Over items

for key,value in student.items():
    print(key,value)
```

```python
### Nested Dictionaries

students = {
    "student1":{'name': 'alien', 'age': 33, 'address': 'india'},
    "student2":{'name': 'alien2', 'age': 25, 'address': 'usa'},
    "student3":{'name': 'alien3', 'age': 32, 'address': 'uk'}  
}

print(students)
```

```python
## Access Nested Dictionaries
print(students["student1"])
print(students["student1"]["name"])

## Iterating over nested Dictionaries

for id,info in students.items():
    print(f"id:{id} with info:{info}")
    for key,value in info.items():
        print(f"{key} with value:{value}")
```

```python
sq_dict = {}
for i in range(10):
    sq_dict[i] = i**2
print(sq_dict)
```

```python
## Dictionary Comprehension

squares = {i:i**2 for i in range(5)}
print(squares)
```

```python
## Conditional dictionary comprehension
squares = {i:i**2 for i in range(10) if i%2==0}
print(squares)
```

```python
## Use a dictionary to count the frequency of elements in list

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency_dct = {}

for num in numbers:
    if num in frequency_dct:
        frequency_dct[num]+=1
    
    else:
        frequency_dct[num] = 1

print(frequency_dct)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}

merged_dict = {**dict1,**dict2}

print(merged_dict)
```

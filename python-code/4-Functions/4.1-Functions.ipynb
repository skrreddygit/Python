```python
## Why Functions?
# Ans: For reusability

num =24

if num%2==0:
    print("the number is even")
else:
    print("Number is odd")
```

```python
def even_or_odd(num):
    """ This Function find out even or odd"""
    if num%2==0:
        print("Number is Even")
    else:
        print("Number is Odd")
```

```python
# Call this function
even_or_odd(7)
```

```python
def add(a,b):
    """ addition of two numbers"""
    return a+b

sum = add(10,20)
print(sum)
```

```python
def greet(name):
    print(f"Hello {name},Welcome to the world")

greet("sai")
#greet()  --> this will throw an error
```

```python
## Default parameters
def greet(name="Guest"):
    print(f"Hello {name},Welcome to the world")

greet()
greet("sai")
```

```python
### Variable Length Arguments
### Positional(*args) and Keyword Arguments

def print_numbers(*sai): #You can use *sai or anything in the place of *args
    for num in sai:
        print(num)

print_numbers(1,2,3,4,5,6,"sai")
```

```python
## Positional Arguments
def print_numbers(*args): #You can use *sai or anything in the place of *args
    for num in args:
        print(num)

print_numbers(1,2,3,4,5,6,"sai")
```

```python
## keyword Arguments (**kwargs)
## All the value in keyword args means value is key and value format ie., dictionary format

def print_details(**kwargs): #You can use *sai or anything in the place of *args
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="alien",age=32,country="india")
```

```python
### Combining positional and keyword arguments


def print_details_new(*args,**kwargs):
    for val in args:
        print(f"Positional argument:{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details_new(1,2,3,4,5,6,"sai",name="alien",age=32,country="india")
```

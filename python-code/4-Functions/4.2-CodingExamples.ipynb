#### Example1: Temperature Conversion

```python
def temp_conversion(temp,unit):
    """ Converts temperature between celsius and Fahrenheit """

    if unit == 'C':
        return temp*9/5 + 32 ## Celsius to fahrenheit

    elif unit == 'F':
        return (temp-32)*5/9
    
    else:
        return None

temp_conversion(25,'C')
temp_conversion(77,'F')
```

#### Example2 : Password Strength Checker

```python
def is_strong_passwd(password):
    """ This function checks if the password is string or not"""

    if len(password) < 8:
        return "Bad password"
    if not any(char.isdigit() for char in password):
        return "Bad password"
    if not any(char.islower() for char in password):
        return "Bad password"
    if not any(char.isupper() for char in password):
        return "Bad password"
    if not any(char in "!@#$%^&*()_+" for char in password):
        return "Bad password"
    
    else:
        return "Good Password"

## calling the function

print(is_strong_passwd("WeakPwd"))
print(is_strong_passwd("StrongPwd1!"))
```

#### Example3: Calculate the Total Cost of Items in a Shopping Cart

```python
def total_cart_cost(cart):
    total_cost = 0

    for item in cart:
        total_cost += item["price"] * item["quantity"]
    
    return total_cost

## Example cart data

cart=[
    {"name":"apple","price":0.5,'quantity':4},
    {"name":"banana","price":0.3,'quantity':6},
    {"name":"Orange","price":0.7,'quantity':3}
]

result = total_cart_cost(cart)
print(result)
```

#### Example4: Check if the given string is palindrome or not

```python
def is_palindrome(word):
    word = word.lower().replace(" ","")
    return word == word[::-1]

print(is_palindrome("A man a plan a Canal Panama"))
print(is_palindrome("malayalam"))
```

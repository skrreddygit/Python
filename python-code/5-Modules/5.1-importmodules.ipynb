#### Importing Modules in Python: Modules and Packages

* In Python,Modules and packages help organize and reuse code.

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi

print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np # aliases

np.array([1,2,3,4,5])
```

```python
from math import * # Importing all functions in math package

print(sqrt(36))
```

```python
from package.maths import add
add(2,3)
```

```python
from package import maths
maths.add(2,3)
```

```python
### Data Serialization

import json
data = {"name": "alien","age":25}

json_str = json.dumps(data)
print(json_str)
print(type(json_str))


parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
### CSV

import csv

with open("example.csv",mode='w',newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["name","age"])
    writer.writerow(["alien",32])

with open("example.csv",mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
## datetime

from datetime import datetime,timedelta

curr_time = datetime.now()
print(curr_time)

yesterday = curr_time - timedelta(days=1)
print(yesterday)
```

```python
## Time
import time

print(time.time())
time.sleep(2)
print(time.time())
```

```python
### regular Expression

import re

pattern = r'\d+'  # \d means digits
text = "there are 123 apples 456"
match = re.search(pattern,text)
print(match)
print(match.group())

```

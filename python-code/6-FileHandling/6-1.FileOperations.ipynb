### File Handling Operations

```python
## Read a Whole File

with open("example.txt",mode='r') as file:
    content = file.read()
    print(content)
```

```python
## Read a File line by line

with open("example.txt",mode="r") as file:
    for line in file:
        print(line)
```

```python
with open("example.txt",mode="r") as file:
    for line in file:
        print(line.strip())  ## .strip() removes the newline character
```

```python
### Writing a File Overwriting , we shouldnt do overwriting

with open("example.txt",mode='w') as file:
    file.write('Hello World\n')
    file.write("this is a New Line")

with open("example.txt",'r') as file:
     for line in file:
        print(line.strip())  ## .strip() removes the newline character
```

```python
### Writing a File ( Without Overwriting)

with open("example.txt",'a') as file:
    file.write("\nAppend operation taking place")

with open("example.txt",'r') as file:
     for line in file:
        print(line.strip())  ## .strip() removes the newline character
```

```python
### Writing List of Files
lines = ["First Line \n","Second Line \n","Third Line \n"]

with open("example.txt",mode='a') as file:
    file.writelines(lines)

```

```python
with open("example.txt",'r') as file:
     for line in file:
        print(line.strip())
```

```python
### Binary Files

#writing to a binary file

data = b'\x00\x01\x02\x03\x04'
with open("example.bin",'wb') as file:  #wb means write bytes
    file.write(data)
```

```python
#reading to a binary file

data = b'\x00\x01\x02\x03\x04'
with open("example.bin",'rb') as file:  #wb means write bytes
    content = file.read()
    print(content)
```

```python
## read the content from source text file and write a destination text file

with open("example.txt",'r') as source_file:
    content = source_file.read()

with open("destination.txt",'w') as destination_file:
    destination_file.write(content)
```

```python
## read a text file and count the number of lines,words,characters
## Counting lines,words,characters in a text filer

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
        return line_count,word_count,char_count
    
file_path = 'example.txt'
lines,words,characters = count_text_file(file_path)

print(f"Lines:{lines},Words:{words},Characters:{characters}")
```

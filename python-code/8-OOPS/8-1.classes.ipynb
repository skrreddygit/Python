```python
### Instance Variables and Methods

class Dog:
    ## constructor
    def __init__(self,name,age): #self is not mandaotory but give some meaningful name but self is used generally
        # instance variables
        self.name = name
        self.age = age

## Create the objects

dog1 = Dog("caeser",3)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
## Define a class with instance methods

class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def bark(self):
        print(f"{self.name} says woof")

dog1 = Dog("caeser",4)
dog1.bark()

dog2 = Dog("buddy",4)
dog2.bark()
```

```python
### Modelling a Bank Account

# Define a class for bank account

class Bank:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self,amount):
        self.balance +=amount
        print(f"New Balance is {self.balance}")

    def withdraw(self,amount):
        if amount>=self.balance:
            print("insufficient funds") 
        else:
            self.balance = self.balance  - amount
            print(f"{amount} is with drawn. New Balance is {self.balance}")

    def get_balance(self):
        return self.balance
```

```python
## create an account

acc1=Bank("alien",500)

print(acc1.get_balance())
acc1.withdraw(1000)
acc1.withdraw(200)
print(acc1.get_balance())
# print(acc1.withdraw(200))
```
